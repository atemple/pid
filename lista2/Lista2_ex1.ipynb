Desenvolva uma função para binarização de imagens que recebe como parâmetro
um limiar. Teste diferentes valores de limiar para binarização de uma imagem em níveis de cinza.

```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt
```

```python
# convert an image with binary 0 and 1
def convert_binary(image_matrix, thresh_val):
    white = 255
    black = 0
    
    initial_conv = np.where((image_matrix <= thresh_val), image_matrix, white)
    final_conv = np.where((initial_conv > thresh_val), initial_conv, black)
    
    return final_conv
```

```python
# get test image
image_aux = data.camera()
```

```python
# convert binary
image1 = convert_binary(image_aux, 120)
```

```python
# show image
plt.imshow(image1, cmap='gray')
```
