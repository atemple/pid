Carregue uma fotografia em níveis de cinza e aplique as diferentes Transformações Não-Lineares discutidas em aula (exponencial, logaritmo, quadrado, raiz quadrada).
Atente-se para manter o resultado no intervalo [0,255].

```python
import numpy as np
import math
from skimage import data
import matplotlib.pyplot as plt
```

```python
# exponential tranformation
def exponential(img):
  a = 255/(np.log(1 + np.max(img)))
  exp_img = a * (np.exp(img) - 1)
  return np.array(exp_img, dtype = np.uint8)
```

```python
# logaritmic tranformation
def logaritmic(img):
  a = 255/(np.log(1 + np.max(img)))
  log_img = a * np.log(1 + img)
  return np.array(log_img, dtype = np.uint8)
```

```python
# square tranformation
def square(img):
  a = 255/(np.log(1 + np.max(img)))
  sqrt_img = a * np.square(img)
  return np.array(sqrt_img, dtype = np.uint8)
```

```python
# sqrt tranformation
def sqrt(img):
  a = 255/(np.log(1 + np.max(img)))
  sqrt_img = a * np.sqrt(img)
  return np.array(sqrt_img, dtype = np.uint8)
```

```python
# get test image
image_aux = data.camera()
```

```python
# apply exponential
image1 = exponential(image_aux)
```

```python
# show image
plt.imshow(image1, cmap='gray')
```

```python
# apply logaritmic
image2 = logaritmic(image_aux)
```

```python
# show image
plt.imshow(image2, cmap='gray')
```

```python
# apply square
image3 = square(image_aux)
```

```python
# show image
plt.imshow(image3, cmap='gray')
```

```python
# apply sqrt
image4 = sqrt(image_aux)
```

```python
# show image
plt.imshow(image4, cmap='gray')
```
