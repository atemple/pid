Implemente uma função de Transformação Não-Linear em dois degraus:
Se f(x,y)<128, aplique função Exponencial; Senão aplique função Logaritmo.

```python
import numpy as np
import math
from skimage import data
import matplotlib.pyplot as plt
```

```python
def exp_log(img):
  a = 255/(np.log(1 + np.max(img)))
  img_n = np.where(img < 128, a * (np.exp(img) - 1), a * np.log(1 + img))
  return np.array(img_n, dtype = np.uint8)
```

```python
# get test image
image_aux = data.camera()
```

```python
# apply exponential
image1 = exp_log(image_aux)
```

```python
# show image
plt.imshow(image1, cmap='gray')
```
