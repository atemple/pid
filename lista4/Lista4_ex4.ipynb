```python
import numpy as np
import math
import random
from skimage import data
import matplotlib.pyplot as plt
import cv2
```

```python
# transform a rgb to gray
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```python
# scale a image with given height and width
def scale(im, nH, nW):
  height = len(im)
  width = len(im[0]) 
  list = [[ im[int(height * r / nH)][int(width * c / nW)]  
             for c in range(nW)] for r in range(nH)]
  return np.asarray(list)
```

```python
# show image 1 and image 2
def show_img(img1, img2):
  fig = plt.figure()
  fig.set_figheight(10)
  fig.set_figwidth(10)

  fig.add_subplot(1,2,1)
  plt.imshow(img1, cmap='gray')

  # display the new image
  fig.add_subplot(1,2,2)
  plt.imshow(img2, cmap='gray')

  plt.show(block=True)
```

```python
# get test image
image_aux = data.camera()
```

```python
# run otsu threshold 
ret, image1 = cv2.threshold(image_aux, 120, 255, cv2.THRESH_OTSU) 
```

```python
# show images original and with otsu threshold
show_img(image_aux, image1)
```
