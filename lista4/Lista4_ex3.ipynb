```python
import numpy as np
import math
import random
from skimage import data
import matplotlib.pyplot as plt
```

```python
# transform a rgb to gray
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
```

```python
# scale a image with given height and width
def scale(im, nH, nW):
  height = len(im)
  width = len(im[0]) 
  list = [[ im[int(height * r / nH)][int(width * c / nW)]  
             for c in range(nW)] for r in range(nH)]
  return np.asarray(list)
```

```python
# show image 1 and image 2
def show_img(img1, img2):
  fig = plt.figure()
  fig.set_figheight(10)
  fig.set_figwidth(10)

  fig.add_subplot(1,2,1)
  plt.imshow(img1, cmap='gray')

  # display the new image
  fig.add_subplot(1,2,2)
  plt.imshow(img2, cmap='gray')

  plt.show(block=True)
```

```python
# global threshold
def global_threshold(img, T):
  m, n = img.shape
  img_n = np.zeros([m, n])

  for i in range(1, m):
      for j in range(1, n):
        pix = img[i,j]
        if (pix <= T[0] and pix > T[1]):
          pix = 255
        elif (pix <= T[1] and pix > T[2]):
          pix = pix * 0.7
        elif (pix <= T[2] and pix > T[3]):
          pix = pix * 0.3
        else:
          pix = 0

        img_n[i,j] = pix

  return img_n.astype(np.uint8)
```

```python
# get test image
image_aux = data.camera()
```

```python
# run global threshold 
image1 = global_threshold(image_aux, [255, 125, 65, 15])
```

```python
# show images original and with global threshold
show_img(image_aux, image1)
```
